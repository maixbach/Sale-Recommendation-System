```python
from bs4 import BeautifulSoup
from bs4 import BeautifulSoup
```

```python
with open("./html_file/full_page_html_t_shirt", "r",  encoding='utf-8') as f:
    html_doc = f.read()
```

```python
# def extract_info(html_doc):
list_item = []
soup = BeautifulSoup(html_doc, 'html.parser')
items = soup.find_all("div", {"class": "col-xs-2-4 shopee-search-item-result__item"})
for item in items:
    item_name = item.find_all("div", {"class": "ie3A+n bM+7UW Cve6sh"})
    if item_name:
        item_info = {
            "name": item_name[0].decode_contents().strip(),
            "price": 0,
            "quantity_sold": 0,
            "shop_name": "",
        }
        item_quantity_sold = item.find_all("div", {"class": "r6HknA uEPGHT"})
        item_link = item.find("a")["href"]
        item_price = item.find_all("div", {"class": "vioxXd rVLWG6"})[0].find_all("span", {"class": "ZEgDH9"})[0].decode_contents().strip()
        if item_quantity_sold:
            item_info["quantity_sold"] = item_quantity_sold[0].decode_contents().strip()
        item_info["price"] = item_price
        item_info["link"] = f"https://shopee.vn{item_link}"
        list_item.append(item_info)
```
